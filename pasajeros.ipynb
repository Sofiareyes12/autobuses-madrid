```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Set 'Etiqueta' column as the index
total.set_index('Etiqueta', inplace=True)
```

```python
# Summary statistics
summary_stats = total.describe()
```

```python
# Bar plot
plt.figure(figsize=(15, 8))
total.groupby('Denominación').sum().plot(kind='bar')
plt.title('Total values for each Denominación across different years')
plt.xlabel('Denominación')
plt.ylabel('Total Value')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
total.groupby('Denominación')['2014'].mean().sort_values(ascending=True).head(20).plot(kind='bar');
```

```python
total.groupby('Denominación')['2014'].mean().sort_values(ascending=False).head(20).plot(kind='bar');
```

```python

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=total.drop('Denominación', axis=1))
plt.title('Distribution of values for each year')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(total.drop('Denominación', axis=1).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation matrix')
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Assuming 'df' is your DataFrame and you want to create a probplot for a specific column '2021'
column_2021 = total['2021']

# Create a probplot
plt.figure(figsize=(8, 6))
stats.probplot(column_2021, dist="norm", plot=plt)
plt.title('Probability Plot for 2021 Data')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```
